```
# list of lists
test = [
    "are you siri?"
]

# columns
columns = ['user1']

testDf = pd.DataFrame(test, columns=columns)

#train_stats = ludwig_model.train(data_csv='Filename.csv')

model = ludwig_model.load('/content/chat_results/_run_3/model')

#results = model.predict(data_csv='results_0/MPG_predictions.csv')
results = model.predict(data_df=testDf)

#print(results)

model.close()
res = results['user2_predictions'].values[0]

import spacy
nlp = spacy.load('en_core_web_sm')
doc = spacy.tokens.doc.Doc(
    nlp.vocab, words=res)
# run the standard pipeline against it
for name, proc in nlp.pipeline:
    doc = proc(doc)
    
print (doc)
```
